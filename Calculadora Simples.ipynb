```python
import math

print("\n******************* Python Calculadora Simples *******************")

def soma(x, y):
    return x + y

def subtração(x, y):
    return x - y

def multiplicação(x, y):
    return x * y

def divisão(x, y):
    if y == 0:
        return "Erro: Divisão por zero não é permitida!"
    return x / y

def porcentagem(x, y):
    return (x * y) / 100

def raiz_Quadrada(x):
    return math.sqrt(x)

print("\nSelecione o número da operação desejada: \n")
print("1 - Soma")
print("2 - Subtração")
print("3 - Multiplicação")
print("4 - Divisão")
print("5 - Porcentagem")
print("6 - Raiz Quadrada")

escolha = input("\nDigite sua opção (1/2/3/4/5/6): ")

try:
    if escolha == '6':
        num = float(input("\nDigite o número para calcular a raiz quadrada: "))
        print("\n")
        print("Raiz quadrada de", num, "=", raiz_Quadrada(num))
        print("\n")

    else:
        num1 = float(input("\nDigite o primeiro número: "))
        num2 = float(input("\nDigite o segundo número: "))

        if escolha == '1':
            print("\n")
            print(num1, "+", num2, "=", soma(num1, num2))
            print("\n")

        elif escolha == '2':
            print("\n")
            print(num1, "-", num2, "=", subtração(num1, num2))
            print("\n")

        elif escolha == '3':
            print("\n")
            print(num1, "*", num2, "=", multiplicação(num1, num2))
            print("\n")

        elif escolha == '4':
            print("\n")
            print(num1, "/", num2, "=", divisão(num1, num2))
            print("\n")
        
        elif escolha == '5':
            print("\n")
            print(num1, "% de", num2, "=", porcentagem(num1, num2))
            print("\n")
        else:
            print("\nOpção Inválida!")

except ValueError:
    print("\nErro: Por favor, insira um número válido.")
```
