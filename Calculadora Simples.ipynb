```python
import math

print("\n******************* Python Calculadora Simples *******************")

## Função para somar dois números
def soma(x, y):
    return x + y

## Função para subtrair o segundo número do primeiro
def subtração(x, y):
    return x - y

## Função para multiplicar dois números
def multiplicação(x, y):
    return x * y

## Função para dividir o primeiro número pelo segundo
## Inclui tratamento de erro para divisão por zero
def divisão(x, y):
    if y == 0:
        return "Erro: Divisão por zero não é permitida!"
    return x / y

## Função para calcular a porcentagem do primeiro número em relação ao segundo
def porcentagem(x, y):
    return (x * y) / 100

## Função para calcular a raiz quadrada de um número
def raiz_Quadrada(x):
    return math.sqrt(x)

print("\nSelecione o número da operação desejada: \n")
print("1 - Soma")
print("2 - Subtração")
print("3 - Multiplicação")
print("4 - Divisão")
print("5 - Porcentagem")
print("6 - Raiz Quadrada")

escolha = input("\nDigite sua opção (1/2/3/4/5/6): ")

try:
    ## Verifica se a operação escolhida é raiz quadrada, que usa apenas um número
    if escolha == '6':
        num = float(input("\nDigite o número para calcular a raiz quadrada: "))
        print("\n")
        print("Raiz quadrada de", num, "=", raiz_Quadrada(num))
        print("\n")

    ## Para as demais operações, solicita dois números do usuário
    else:
        num1 = float(input("\nDigite o primeiro número: "))
        num2 = float(input("\nDigite o segundo número: "))

        ## Executa a operação de acordo com a escolha do usuário
        if escolha == '1':
            print("\n")
            print(num1, "+", num2, "=", soma(num1, num2))
            print("\n")

        elif escolha == '2':
            print("\n")
            print(num1, "-", num2, "=", subtração(num1, num2))
            print("\n")

        elif escolha == '3':
            print("\n")
            print(num1, "*", num2, "=", multiplicação(num1, num2))
            print("\n")

        elif escolha == '4':
            print("\n")
            print(num1, "/", num2, "=", divisão(num1, num2))
            print("\n")
        
        elif escolha == '5':
            print("\n")
            print(num1, "% de", num2, "=", porcentagem(num1, num2))
            print("\n")

        ## Tratamento de entrada inválida para escolha de operação
        else:
            print("\nOpção Inválida!")

## Tratamento de erro para entradas que não sejam números
except ValueError:
    print("\nErro: Por favor, insira um número válido.")
```
